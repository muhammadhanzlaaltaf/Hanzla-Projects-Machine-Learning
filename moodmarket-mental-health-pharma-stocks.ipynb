# **6.0 Data Cleaning & Text Preprocessing**

This section is where the data actually becomes usable. Raw social media text is noisy — full of URLs, punctuation, inconsistent casing, and HTML artifacts. We strip all of that away before anything else touches the text.

## 6.1 Standardise Column Names

```python
# Rename to consistent snake_case
rename_map = {}
for col in df.columns:
    low = col.lower().strip()
    if 'text' in low or 'statement' in low or 'post' in low:
        rename_map[col] = 'text'
    elif 'label' in low or 'status' in low or 'class' in low:
        rename_map[col] = 'label'
    else:
        rename_map[col] = low.replace(' ', '_')

df = df.rename(columns=rename_map)
print(f"✅ Columns renamed: {df.columns.tolist()}")
```

## 6.2 Drop Nulls in Core Columns

```python
before = len(df)
df = df.dropna(subset=['text', 'label'])
df['text']  = df['text'].astype(str).str.strip()
df['label'] = df['label'].astype(str).str.strip()
print(f"✅ Rows before: {before:,}  →  after: {len(df):,}  (removed {before - len(df)} nulls)")
```

## 6.3 Standardise Label Names

```python
# Clean and title-case all labels
df['label'] = df['label'].str.strip().str.title()
print("📋 Label distribution after cleaning:")
print(df['label'].value_counts())
```

## 6.4 Text Cleaning Pipeline

Four operations: lowercase everything, strip URLs, remove HTML tags, and kill non-alphabetic characters. Simple, but it removes a surprising amount of noise. We also compute text length and word count — both turn out to be useful EDA variables.

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+', '', text)        # remove URLs
    text = re.sub(r'<.*?>', '', text)                   # remove HTML tags
    text = re.sub(r'[^a-z\s]', ' ', text)              # remove non-alpha
    text = re.sub(r'\s+', ' ', text).strip()           # collapse whitespace
    return text

df['text_clean'] = df['text'].apply(clean_text)
df['text_len']   = df['text_clean'].apply(len)
df['word_count'] = df['text_clean'].apply(lambda x: len(x.split()))

print("✅ Text cleaning complete.")
print(f"   Avg text length  : {df['text_len'].mean():.0f} chars")
print(f"   Avg word count   : {df['word_count'].mean():.1f} words")
df[['text', 'text_clean', 'text_len', 'word_count']].head(3)
```

## 6.5 Create Binary Distress Flag

Seven labels is a lot for a binary classification target. We collapse them into a single `is_distress` flag — Normal = 0, everything else = 1. This feeds directly into our MHDI aggregation.

```python
# Mental health distress binary: 0 = Normal, 1 = Distress
DISTRESS_LABELS = ['Depression', 'Suicidal', 'Anxiety', 'Stress', 'Bi-Polar', 'Personality Disorder']
df['is_distress'] = df['label'].apply(lambda x: 1 if x in DISTRESS_LABELS else 0)

print("✅ Binary distress column created.")
print(f"   Distress (1) : {df['is_distress'].sum():,}  ({df['is_distress'].mean()*100:.1f}%)")
print(f"   Normal   (0) : {(df['is_distress']==0).sum():,}  ({(1-df['is_distress'].mean())*100:.1f}%)")
```

## 7.6 TF-IDF Most Distinctive Terms

```python
# Fit TF-IDF on a sample for speed, then show per-class top terms
tfidf = TfidfVectorizer(max_features=5000, stop_words='english', ngram_range=(1,2))
tfidf.fit(df['text_clean'])

fig, axes = plt.subplots(2, 4, figsize=(16, 7))
axes = axes.flatten()
feature_names = tfidf.get_feature_names_out()
labels_sorted = df['label'].value_counts().index.tolist()

for i, label in enumerate(labels_sorted):
    ax = axes[i]
    subset = df[df['label'] == label]['text_clean']
    X      = tfidf.transform(subset)
    scores = np.asarray(X.mean(axis=0)).flatten()
    top_idx = scores.argsort()[-10:][::-1]
    top_terms  = [feature_names[j] for j in top_idx]
    top_scores = scores[top_idx]
    shades = plt.cm.Blues_r(np.linspace(0.2, 0.8, 10))
    ax.barh(top_terms[::-1], top_scores[::-1], color=shades, edgecolor='none', height=0.6)
    ax.set_title(label, fontsize=11, fontweight='bold', pad=6)
    ax.set_xlabel('Avg TF-IDF', fontsize=9)
    sns.despine(ax=ax)

if len(labels_sorted) < len(axes):
    for j in range(len(labels_sorted), len(axes)):
        axes[j].axis('off')

plt.suptitle('Top TF-IDF Bigrams & Terms per Category', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

# **8.0 Mental Health Discourse Index (MHDI)**

This is the creative core of the project. The Kaggle dataset is static — no dates, no time dimension. But stock prices live in time. So we have to build the bridge ourselves.

## 8.1 Building the MHDI Time-Series

> **Design Note:** The Kaggle dataset has no timestamps — it's a snapshot collection, not a live feed. To build the time-series bridge, I construct a **Mental Health Discourse Index (MHDI)** by: (1) computing per-label distress proportions from the static dataset; (2) simulating a daily discourse signal over Jan 2020–Dec 2023 that respects COVID-era spikes; (3) adding weekly seasonality and bounded noise to mirror real Reddit posting rhythms. This is a transparent proxy — the v2 roadmap covers how to replace it with live Reddit PRAW data.

```python
np.random.seed(42)
START_DATE = '2020-01-01'
END_DATE   = '2023-12-31'
dates      = pd.date_range(START_DATE, END_DATE, freq='B')  # business days only

# Real class proportions from dataset
label_props = df['label'].value_counts(normalize=True).to_dict()
distress_base = df['is_distress'].mean()

# Build MHDI with autocorrelated noise + seasonality
n = len(dates)
trend    = np.linspace(0, 0.05, n)
seasonal = 0.04 * np.sin(2 * np.pi * np.arange(n) / 252)  # annual cycle
noise    = np.random.normal(0, 0.02, n)

# Autocorrelated noise (AR(1))
ar_noise = np.zeros(n)
for t in range(1, n):
    ar_noise[t] = 0.7 * ar_noise[t-1] + noise[t]

mhdi = distress_base + trend + seasonal + ar_noise * 0.5
mhdi = np.clip(mhdi, 0.3, 0.98)

# Post volume (higher on weekdays, spikes for major events)
base_volume = np.random.poisson(lam=320, size=n)
# Simulate COVID-19 spike (March 2020 onwards)
covid_mask = (dates >= '2020-03-01') & (dates <= '2020-12-31')
base_volume[covid_mask] += np.random.randint(100, 400, covid_mask.sum())

time_df = pd.DataFrame({
    'Date'        : dates,
    'mhdi'        : mhdi,
    'post_volume' : base_volume,
    'weekday'     : dates.dayofweek
})

print(f"✅ MHDI time-series created: {len(time_df):,} trading days")
print(f"   Date range : {time_df['Date'].min().date()} → {time_df['Date'].max().date()}")
print(f"   Avg MHDI   : {time_df['mhdi'].mean():.4f}")
time_df.head()
```

```python
TICKERS = ['PFE', 'LLY', 'ABBV', 'MRK']

print("📡 Fetching pharma stock data from Yahoo Finance...")
raw = yf.download(TICKERS, start=START_DATE, end=END_DATE,
                  auto_adjust=True, progress=False)

stock_df = raw.stack(level=1).rename_axis(['Date','Ticker']).reset_index()
stock_df  = stock_df.sort_values(['Ticker','Date'])

# Daily return and next-day direction target
stock_df['daily_return'] = stock_df.groupby('Ticker')['Close'].pct_change()
stock_df['target']       = (stock_df.groupby('Ticker')['Close'].shift(-1)
                            > stock_df['Close']).astype(int)
stock_df = stock_df.dropna()

print(f"✅ Stock data loaded: {stock_df.shape[0]:,} rows")
print(f"   Tickers : {stock_df['Ticker'].unique().tolist()}")
stock_df.head()
```

## 8.3 Merge MHDI + Stock Data & Feature Engineering

Once MHDI and stock prices share a common date index, we build lag features, rolling averages, and volatility windows — the signals the model actually learns from.

```python
# Merge on date
merged_df = pd.merge(stock_df, time_df, on='Date', how='inner')

# ── Feature engineering ────────────────────────────────────────
merged_df = merged_df.sort_values(['Ticker','Date'])

# Lag MHDI features
for lag in [1, 2, 3, 5]:
    merged_df[f'mhdi_lag{lag}']   = merged_df.groupby('Ticker')['mhdi'].shift(lag)
    merged_df[f'return_lag{lag}'] = merged_df.groupby('Ticker')['daily_return'].shift(lag)

# Rolling metrics
merged_df['mhdi_7d_avg']     = merged_df.groupby('Ticker')['mhdi'].transform(
    lambda x: x.rolling(7, min_periods=3).mean())
merged_df['mhdi_7d_std']     = merged_df.groupby('Ticker')['mhdi'].transform(
    lambda x: x.rolling(7, min_periods=3).std())
merged_df['volatility_14d']  = merged_df.groupby('Ticker')['daily_return'].transform(
    lambda x: x.rolling(14, min_periods=5).std())
merged_df['volume_spike']    = merged_df.groupby('Ticker')['post_volume'].transform(
    lambda x: x / x.rolling(7, min_periods=3).mean())

# Time features
merged_df['DayOfWeek']  = merged_df['Date'].dt.dayofweek
merged_df['Month']      = merged_df['Date'].dt.month
merged_df['Quarter']    = merged_df['Date'].dt.quarter
merged_df['Is_Weekend'] = (merged_df['DayOfWeek'] >= 5).astype(int)

# Encode ticker
le_ticker = LabelEncoder()
merged_df['Ticker_enc'] = le_ticker.fit_transform(merged_df['Ticker'])

merged_df = merged_df.dropna()
print(f"✅ Merged dataset: {merged_df.shape[0]:,} rows × {merged_df.shape[1]} columns")
print(f"   Class balance — UP: {merged_df['target'].mean()*100:.1f}% | DOWN: {(1-merged_df['target'].mean())*100:.1f}%")
merged_df.head()
```

# **10.0 Machine Learning — Predicting Next-Day Stock Direction**

Here's where the rubber meets the road. Binary classification: given everything we know today (MHDI, lag features, rolling volatility), will the stock close *higher* tomorrow?

Plan: benchmark 7 classical models → tune the best → then run an **LSTM neural network** to see whether sequence modelling improves on the tree-based approaches.

## 10.1 Feature Selection & Target Variable

18 features. Target is binary: 1 if tomorrow's close > today's close, 0 otherwise. No future data — strictly what you'd know at end of day.

```python
features = [
    'mhdi', 'post_volume', 'daily_return', 'volatility_14d', 'volume_spike',
    'mhdi_lag1', 'mhdi_lag2', 'mhdi_lag3', 'mhdi_lag5',
    'mhdi_7d_avg', 'mhdi_7d_std',
    'return_lag1', 'return_lag2', 'return_lag3',
    'DayOfWeek', 'Month', 'Quarter', 'Ticker_enc'
]

target = 'target'

X = merged_df[features].copy()
y = merged_df[target].copy()

print(f"✅ Features selected  : {X.shape[1]}")
print(f"✅ Target             : {target} (1=UP, 0=DOWN)")
print(f"✅ Dataset shape      : {X.shape}")
print(f"✅ Positive class %   : {y.mean()*100:.1f}%")
```

## 10.2 Time-Aware Train / Test Split

Critical detail: you cannot randomly shuffle financial time-series. Doing so creates data leakage — future observations bleed into training. We sort by date, take first 80% as training, last 20% as test.

```python
# Time-aware split — no data leakage
merged_sorted = merged_df.sort_values(['Date','Ticker'])
split_idx = int(len(merged_sorted) * 0.8)

X_train = merged_sorted[features].iloc[:split_idx]
X_test  = merged_sorted[features].iloc[split_idx:]
y_train = merged_sorted[target].iloc[:split_idx]
y_test  = merged_sorted[target].iloc[split_idx:]

print(f"✅ Train set   : {X_train.shape[0]:,} rows  (up to {merged_sorted['Date'].iloc[split_idx-1].date()})")
print(f"✅ Test set    : {X_test.shape[0]:,} rows  (from {merged_sorted['Date'].iloc[split_idx].date()})")
print(f"   Train UP%  : {y_train.mean()*100:.1f}%")
print(f"   Test  UP%  : {y_test.mean()*100:.1f}%")
```

## 10.3 Feature Scaling

Tree models don't need scaling, but neural networks and distance-based models do. We apply StandardScaler and keep both raw and scaled arrays.

```python
scaler     = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)
print("✅ Features scaled with StandardScaler")
```

## 10.4 Define the Model Lineup

Seven models — from interpretable baselines (Logistic Regression, Decision Tree) to ensemble powerhouses (Random Forest, XGBoost). We want to understand how much complexity actually buys us.

```python
models = {
    'Logistic Regression'   : LogisticRegression(max_iter=1000, random_state=42),
    'KNN Classifier'        : KNeighborsClassifier(n_neighbors=9),
    'Decision Tree'         : DecisionTreeClassifier(max_depth=6, random_state=42),
    'Random Forest'         : RandomForestClassifier(n_estimators=200, random_state=42),
    'Gradient Boosting'     : GradientBoostingClassifier(n_estimators=200, random_state=42),
    'AdaBoost'              : AdaBoostClassifier(n_estimators=200, random_state=42),
    'XGBoost'               : XGBClassifier(n_estimators=200, random_state=42,
                                             eval_metric='logloss', verbosity=0),
}
print(f"✅ {len(models)} models ready for benchmarking")
```

## 10.5 Benchmark All 7 Classical Models

Train each model, measure AUC-ROC, F1, Accuracy, Precision, and Recall. We're targeting AUC in the 0.52–0.65 range — stock direction prediction is genuinely hard, and anything consistently above 0.60 is real signal.

```python
SCALED = {'Logistic Regression', 'KNN Classifier'}
results = []

for name, model in models.items():
    Xtr = X_train_sc if name in SCALED else X_train
    Xte = X_test_sc  if name in SCALED else X_test

    model.fit(Xtr, y_train)
    y_pred = model.predict(Xte)
    y_prob = model.predict_proba(Xte)[:,1] if hasattr(model,'predict_proba') else None

    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec  = recall_score(y_test, y_pred, zero_division=0)
    f1   = f1_score(y_test, y_pred, zero_division=0)
    auc  = roc_auc_score(y_test, y_prob) if y_prob is not None else np.nan

    results.append({'Model':name,'Accuracy':acc,'Precision':prec,
                    'Recall':rec,'F1 Score':f1,'AUC-ROC':auc})
    print(f"  {name:25} → Acc: {acc:.4f} | F1: {f1:.4f} | AUC: {auc:.4f}")

results_df = pd.DataFrame(results).sort_values('AUC-ROC', ascending=False)
print("\n✅ Benchmarking complete!")
```

# **10.14 Neural Network — LSTM for Temporal Sequence Modelling**

Classical models treat each day as an independent observation. But stock markets have *memory* — the pattern of MHDI scores over the past week contains information that a single-day snapshot misses. That's exactly where LSTMs shine.

An **LSTM (Long Short-Term Memory)** network is a type of recurrent neural network designed to learn patterns across time sequences. Instead of feeding it one day's features at a time, we construct rolling windows of the past 10 trading days and let the LSTM find whatever temporal structure exists in the MHDI signal.

This won't magically solve stock prediction — but it's a meaningful architectural upgrade over the tree-based benchmark, and the comparison is genuinely interesting.

## 10.14.1 Install & Import Deep Learning Libraries

```python
# Deep Learning — TensorFlow/Keras for LSTM
# Run this if tensorflow is not already installed:
# !pip install tensorflow --quiet

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (LSTM, Dense, Dropout, BatchNormalization,
                                      Bidirectional, Input)
from tensorflow.keras.callbacks import (EarlyStopping, ReduceLROnPlateau,
                                         ModelCheckpoint)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
import tensorflow.keras.backend as K

print(f"✅ TensorFlow version: {tf.__version__}")
print(f"   GPU available: {len(tf.config.list_physical_devices('GPU')) > 0}")
```

## 10.14.2 Build Sequence Windows

We reshape the feature matrix into 3D windows of shape `(samples, timesteps, features)`. Each sample represents 10 consecutive trading days, and the target is whether day 11 will close higher.

Using a 10-day lookback window — roughly two trading weeks — which is long enough to capture weekly patterns in the MHDI without being so long that the earliest observations become irrelevant.

```python
# ── LSTM Sequence Window Builder ───────────────────────────────────────
LOOKBACK = 10  # 10 trading days per sequence window

def make_sequences(X_arr, y_arr, lookback=LOOKBACK):
    """Convert a 2D feature array into 3D (samples, timesteps, features) windows."""
    Xs, ys = [], []
    for i in range(lookback, len(X_arr)):
        Xs.append(X_arr[i - lookback : i])
        ys.append(y_arr[i])
    return np.array(Xs), np.array(ys)

# Use scaled features for the LSTM (neural nets need normalised inputs)
X_seq, y_seq = make_sequences(X_train_sc, y_train.values)
X_seq_test, y_seq_test = make_sequences(X_test_sc, y_test.values)

print(f"✅ Sequence arrays constructed")
print(f"   Train sequences : {X_seq.shape}   → (samples, timesteps, features)")
print(f"   Test  sequences : {X_seq_test.shape}")
print(f"   Class balance   — Train UP: {y_seq.mean()*100:.1f}% | Test UP: {y_seq_test.mean()*100:.1f}%")
```

## 10.14.3 Build the Bidirectional LSTM

Architecture choices:
- **Bidirectional LSTM** — reads the sequence both forward and backward, capturing context from both directions.
- **Batch Normalisation** — keeps activations stable, especially helpful with financial time-series which can have sudden volatility spikes.
- **Dropout (0.3)** — regularisation to prevent overfitting on noisy financial data.
- **L2 regularisation** on the Dense layers — another layer of overfitting defence.
- **Sigmoid output** — binary classification probability.

```python
# ── Bidirectional LSTM Architecture ────────────────────────────────────
n_features = X_seq.shape[2]

def build_lstm_model(n_features, lookback, lr=1e-3):
    model = Sequential([
        Input(shape=(lookback, n_features)),

        # First Bi-LSTM layer — extract rich temporal representations
        Bidirectional(LSTM(64, return_sequences=True,
                           kernel_regularizer=l2(1e-4))),
        BatchNormalization(),
        Dropout(0.3),

        # Second Bi-LSTM layer — compress into a single context vector
        Bidirectional(LSTM(32, return_sequences=False,
                           kernel_regularizer=l2(1e-4))),
        BatchNormalization(),
        Dropout(0.25),

        # Dense head
        Dense(32, activation='relu', kernel_regularizer=l2(1e-4)),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dense(1, activation='sigmoid')   # Binary output
    ])

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
    )
    return model

lstm_model = build_lstm_model(n_features, LOOKBACK)
lstm_model.summary()
```

## 10.14.4 Train the LSTM

Three callbacks keep training honest:
- **EarlyStopping** — stops when validation AUC stops improving, restores the best weights automatically.
- **ReduceLROnPlateau** — halves the learning rate if val_loss stagnates for 5 epochs.
- **ModelCheckpoint** — saves the best model weights to disk.

```python
# ── Callbacks ───────────────────────────────────────────────────────────
callbacks = [
    EarlyStopping(
        monitor='val_auc', mode='max',
        patience=12, restore_best_weights=True, verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss', factor=0.5,
        patience=5, min_lr=1e-6, verbose=1
    ),
]

# ── Train ────────────────────────────────────────────────────────────────
print("🚀 Training Bidirectional LSTM...")
history = lstm_model.fit(
    X_seq, y_seq,
    validation_split=0.15,
    epochs=80,
    batch_size=64,
    callbacks=callbacks,
    verbose=1,
    class_weight={0: 1.0, 1: y_seq.mean() / (1 - y_seq.mean())}  # Handle class imbalance
)

print(f"\n✅ Training complete — stopped at epoch {len(history.history['loss'])}.")
```

## 10.14.5 Training History — Loss & AUC Curves

A healthy training curve shows validation loss tracking closely to training loss, with AUC steadily improving before plateauing. A large gap between training and validation signals overfitting.

```python
# ── Training Curves ─────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss
axes[0].plot(history.history['loss'],     color='#1e3a5f', linewidth=2,   label='Train Loss')
axes[0].plot(history.history['val_loss'], color='#e85d04', linewidth=2,   label='Val Loss', linestyle='--')
axes[0].set_title('Training vs Validation Loss', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Binary Crossentropy')
axes[0].legend(); sns.despine(ax=axes[0])

# AUC
axes[1].plot(history.history['auc'],     color='#1e3a5f', linewidth=2,   label='Train AUC')
axes[1].plot(history.history['val_auc'], color='#e85d04', linewidth=2,   label='Val AUC', linestyle='--')
axes[1].axhline(0.5, color='#aaa', linewidth=1, linestyle=':', label='Random Baseline')
axes[1].set_title('Training vs Validation AUC', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('AUC-ROC')
axes[1].legend(); sns.despine(ax=axes[1])

plt.suptitle('LSTM Training History', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

Final evaluation on the held-out test sequences. We compute the same metrics as the classical models so we can do a direct apples-to-apples comparison.

```python
# ── LSTM Test Evaluation ────────────────────────────────────────────────
lstm_prob = lstm_model.predict(X_seq_test, verbose=0).flatten()
lstm_pred = (lstm_prob >= 0.5).astype(int)

lstm_auc  = roc_auc_score(y_seq_test, lstm_prob)
lstm_f1   = f1_score(y_seq_test, lstm_pred, zero_division=0)
lstm_acc  = accuracy_score(y_seq_test, lstm_pred)
lstm_prec = precision_score(y_seq_test, lstm_pred, zero_division=0)
lstm_rec  = recall_score(y_seq_test, lstm_pred, zero_division=0)

print("=" * 55)
print("     LSTM (Bidirectional) — TEST SET PERFORMANCE")
print("=" * 55)
print(f"  AUC-ROC   : {lstm_auc:.4f}")
print(f"  F1 Score  : {lstm_f1:.4f}")
print(f"  Accuracy  : {lstm_acc:.4f}")
print(f"  Precision : {lstm_prec:.4f}")
print(f"  Recall    : {lstm_rec:.4f}")
print("=" * 55)

# Compare with XGBoost
print(f"\n  XGBoost (tuned) AUC : {auc_t:.4f}")
print(f"  LSTM AUC            : {lstm_auc:.4f}")
delta = lstm_auc - auc_t
if delta > 0:
    print(f"  ✅ LSTM improves AUC by +{delta:.4f} — temporal modelling adds value")
else:
    print(f"  ⚠️  LSTM AUC is {abs(delta):.4f} lower — XGBoost remains champion on this window")
    print(f"     (This is common with limited financial data; LSTM shines with longer histories)")
```

# **11.0 Prediction Error Analysis**

Not all wrong predictions are equal. A confidently wrong model is worse than an uncertain one. This section digs into the probability distributions and error patterns behind our predictions.

## 11.1 Predicted Probability Distribution by True Class

If the model is working, the probability distribution for actual UP days should skew right, and for actual DOWN days it should skew left. Heavy overlap in the middle is the uncertain zone — and that's fine, as long as the confident predictions are correct.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for label, color, name in [(0,'#e85d04','Actual DOWN'),(1,'#1e3a5f','Actual UP')]:
    axes[0].hist(y_prob_best[y_test==label], bins=25, color=color,
                 alpha=0.65, edgecolor='white', label=name, linewidth=0.5)
axes[0].axvline(0.5, color='#aaa', linestyle='--', linewidth=1.2)
axes[0].set_title('Predicted Probability by True Class', fontsize=12, fontweight='bold', pad=8)
axes[0].set_xlabel('P(Price UP)', fontsize=10)
axes[0].set_ylabel('Frequency', fontsize=10)
axes[0].legend(fontsize=10)
sns.despine(ax=axes[0])

# Calibration curve
prob_bins  = np.linspace(0, 1, 11)
bin_center = (prob_bins[:-1] + prob_bins[1:]) / 2
bin_acc    = []
for lo, hi in zip(prob_bins[:-1], prob_bins[1:]):
    mask = (y_prob_best >= lo) & (y_prob_best < hi)
    bin_acc.append(y_test[mask].mean() if mask.sum() > 0 else np.nan)

axes[1].plot(bin_center, bin_center, '--', color='#aaa', linewidth=1.2, label='Perfect calibration')
axes[1].plot(bin_center, bin_acc, 'o-', color='#1e3a5f', linewidth=1.8,
             markersize=7, label='XGBoost')
axes[1].set_title('Calibration Curve', fontsize=12, fontweight='bold', pad=8)
axes[1].set_xlabel('Mean Predicted Probability', fontsize=10)
axes[1].set_ylabel('Fraction of Positives', fontsize=10)
axes[1].legend(fontsize=10)
sns.despine(ax=axes[1])

plt.suptitle('Prediction Error Analysis', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
